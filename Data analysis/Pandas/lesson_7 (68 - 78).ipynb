## 二、数据转换

### 移除重复数据

```python
data = pd.DataFrame({'k1':['one', 'two'] * 3 + ['two'],
                    'k2':[1,1,2,3,3,4,4]})
data
```

```python
# 检查数据是否重复 返回布尔类型
data.duplicated()
```

```python
# 会把重复的数据删除后返回新的对象，默认保留第一次出现的数据
data.drop_duplicates()
```

```python
data['v1'] = np.arange(7)
data
```

```python
# 按照某一列内的数据是否重复进行过滤
data.drop_duplicates(['k1'])
```

```python
# 保留最后一个重复的数据
data.drop_duplicates(['k1'], keep='last')
```

### 利用函数或映射进行数据转换

```python
data = pd.DataFrame({"food":["Apple", "banana", 'orange', 'apple', 'Mango', 'tomato'],
                    'price':[4, 3, 3.5, 6, 12, 3]})
data
```

```python
meat = {'apple':'fruit', 
        'banana':'fruit',
       'orange':'fruit',
       'mango':'fruit',
       'tomato':'vegetable'}
```

```python
# 把字符串转化成全部小写格式
low = data['food'].str.lower()
low
```

```python
# 映射
data['class'] = data['food'].str.lower().map(meat)
data
```

```python
# 函数
data['class1'] = data['food'].map(lambda x: meat[x.lower()])
data
```

### 重命名轴索引

```python
data = pd.DataFrame(np.arange(12).reshape(3,4),
                   index=['BeiJing', 'Tokyo', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
data
```

```python
# 重新索引reindex 产生新的对象
data.reindex(['Tokyo', 'New York', 'BeiJing'])
```

```python
data
```

```python
# 全大写
tran = lambda x: x[:4].upper()
data.index.map(tran)
```

```python
# 在原索引的基础上进行简单修改，不会对数据值产生影响
data.index = data.index.map(tran)
data
```

```python
col = lambda x: x[:5].title()
data.columns = data.columns.map(col)
```

```python
data
```

```python
# rename 更新索引格式
data.rename(index=str.title, columns=str.upper)
```

```python
# 结合字典型对象对标签进行更改
data.rename(index={'BEIJ':"北京", 'TOKY':'东京', 'NEW ':'纽约'}, columns={'One':1, 'Two':2, 'Three':3, 'Four':4})
```

```python
data
```

```python
# 使用inplace参数可以就地修改
data.rename(index={'BEIJ':"北京", 'TOKY':'东京', 'NEW ':'纽约'}, 
            columns={'One':1, 'Two':2, 'Three':3, 'Four':4}, 
            inplace=True)
data
```

### 离散化和面元划分 ------ 分阶段

### 正则表达式

```python
import re
```

```python
text = 'foo  bar\t bat  \tqq'
text
```

```python
re.split('\s+', text)
```

```python
# re.compile() 将正则表达式编译成一个对象，以后也可以直接使用
res = re.compile('\s+')
res.split(text)
```

```python
reg = re.split(res, text)
reg
```

```python
# .findall() 查找匹配所有
res.findall(text)
```

```python
re.findall(res, text)
```

```python
# 替换
re.sub(res, '9', text)
```

```python
res.sub('9', text)
```

```python
# 只匹配起始字符
t1 = re.match('b', text)
print(t1)
```

```python
# 匹配全体字符，但只要找到一个就返回匹配项对象，后面还有没有也不管啦
t2 = re.search('b', text)
print(t2.group())
```
